```python
import numpy as np
import pandas as pd
```

```python
xs = np.linspace(-6, 6, 100)
ys = np.sinc(xs)
```

```python
df = pd.DataFrame({'x': xs, 'sinc':ys})
df.plot.line(x='x', y='sinc')
```

```python
import matplotlib.pyplot as plt
```
